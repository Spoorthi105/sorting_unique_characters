```python
import tkinter as tk
import calendar

class CalendarApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Python Calendar App")

        self.date = tk.StringVar()
        self.month_year_label = tk.Label(root, textvariable=self.date, font=("Helvetica", 16))
        self.month_year_label.pack()

        self.frame = tk.Frame(root)
        self.frame.pack()

        self.cal = calendar.Calendar()
        self.create_calendar()

        self.prev_button = tk.Button(root, text="Previous Month", command=self.prev_month)
        self.prev_button.pack(side=tk.LEFT)

        self.next_button = tk.Button(root, text="Next Month", command=self.next_month)
        self.next_button.pack(side=tk.RIGHT)

    def create_calendar(self):
        self.current_year = calendar.datetime.datetime.now().year
        self.current_month = calendar.datetime.datetime.now().month

        self.update_calendar()

    def update_calendar(self):
        cal_text = self.cal.formatmonth(self.current_year, self.current_month)
        self.date.set(calendar.month_name[self.current_month] + " " + str(self.current_year))
        if hasattr(self, "cal_label"):
            self.cal_label.destroy()
        self.cal_label = tk.Label(self.frame, text=cal_text, font=("Helvetica", 14))
        self.cal_label.pack()

    def prev_month(self):
        self.current_month -= 1
        if self.current_month == 0:
            self.current_month = 12
            self.current_year -= 1
        self.update_calendar()

    def next_month(self):
        self.current_month += 1
        if self.current_month == 13:
            self.current_month = 1
            self.current_year += 1
        self.update_calendar()

if __name__ == "__main__":
    root = tk.Tk()
    app = CalendarApp(root)
    root.mainloop()





```
